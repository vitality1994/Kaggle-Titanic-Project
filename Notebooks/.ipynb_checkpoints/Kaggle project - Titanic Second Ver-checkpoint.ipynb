### b. Eliminate Outliers
I will use Tukey's fences to identify outliers in each feature.

Tukey's fences:<br>
Outliers are defined outside of the interval [Q1-k(Q3-Q1), Q3+k(Q3-Q1)] when k=1.5. <br>

```python
# From here: https://www.kaggle.com/yassineghouzam/titanic-top-4-with-ensemble-modeling
def detect_outliers(df,n,features):
    """
    Takes a dataframe df of features and returns a list of the indices
    corresponding to the observations containing more than n outliers according
    to the Tukey method.
    """
    outlier_indices = []
    
    # iterate over features(columns)
    for col in features:
        # 1st quartile (25%)
        Q1 = np.percentile(df[col], 25)
        # 3rd quartile (75%)
        Q3 = np.percentile(df[col],75)
        # Interquartile range (IQR)
        IQR = Q3 - Q1
        
        # outlier step
        outlier_step = 1.5 * IQR
        
        # Determine a list of indices of outliers for feature col
        outlier_list_col = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step )].index
        
        # append the found outlier indices for col to the list of outlier indices 
        outlier_indices.extend(outlier_list_col)
        
    # select observations containing more than 2 outliers
    outlier_indices = Counter(outlier_indices) 
    multiple_outliers = list( k for k, v in outlier_indices.items() if v > n )
    return multiple_outliers   

# detect outliers from Age, SibSp , Parch and Fare
Outliers_to_drop = detect_outliers(train_data,2,["Age","SibSp","Parch","Fare"])
```

## Feature engineering
1. Change Name data to categorical & nuemrical data: <br>
There might be the relationship between titles of passengers and the survival rate.<br><br>
2. Make two categories(Kids(0-6), childeren&adolescent&young adults(20-30), adults&seniors(30-)) of age data<br> because subpopulations have some relationship with the survival rate.<br><br>
3. Change Sex data to Categorical&Numeric data.

 ### 1. Change Name data to categorical&numerical data

```python
import re
# import 'regular expression' module to extract 'title' from name data.

p = re.compile('[a-zA-Z]+[.]')
# Regular expressions: alphabets + .(dot)
p.findall(train_data['Name'][0])
# The name for the first raw is 'Braud, Mr. Owen Harris'.
# Therefore, the code above prints 'Mr.'

rank_name_list = []
for i in range(len(train_data['Name'])):
    rank_name_list.append(p.findall(train_data['Name'][i]))

rank_name = pd.DataFrame(rank_name_list)
rank_name = rank_name.drop([1], axis=1)
# Drop the unnecessary column
rank_name=rank_name.rename(columns={0:'Title'})
# Change a name of the column.
train_data=pd.concat([train_data, rank_name], axis=1)
# Combine with the original train_data (Dataframe)\

# Categorized the titles to reduce the number of titles.
train_data['Title'] = train_data['Title'].replace(['Lady.', 'Countess.','Capt.', 'Col.','Don.', 
                                              'Dr.', 'Major.', 'Rev.', 'Sir.', 'Jonkheer.', 
                                              'Dona.'], 'Rare.')

train_data['Title'] = train_data['Title'].replace('Mlle.', 'Miss.')
train_data['Title'] = train_data['Title'].replace('Ms.', 'Miss.')
train_data['Title'] = train_data['Title'].replace('Mme.', 'Mrs.')

# Change title data to numeric data.
train_data['Title'] = train_data['Title'].replace('Mr.', 5)
train_data['Title'] = train_data['Title'].replace('Mrs.', 1)
train_data['Title'] = train_data['Title'].replace('Miss.', 2)
train_data['Title'] = train_data['Title'].replace('Master.', 3)
train_data['Title'] = train_data['Title'].replace('Rare.', 4)

train_data=train_data.drop('Name', axis=1)
```

### 2. Make three categories(Kids(0-6), childeren&adolescents&young adults(7-30), adults&seniors(30-)) of age data<br>

```python
# 0~6 kids, 6~30 group2, 20~30 childerens & adolescents& young adults, 
# 30~ adults&seniors
age_group_list=[]
for i in range(len(train_data['Age'])):
    if 0<=train_data['Age'][i] and train_data['Age'][i]<=6:
        age_group_list.append(3)
    elif 6<train_data['Age'][i] and train_data['Age'][i]<=30:
        age_group_list.append(1)
    elif 30<train_data['Age'][i]:
        age_group_list.append(2)
        
age_group = pd.DataFrame(age_group_list, columns=['age_group'])

train_data = pd.concat([train_data, age_group], axis=1)
# 'concat' function is for combining two different dataframes in terms of the index
# when we combine those horizontally. (axis=1)

train_data=train_data.drop('Age', axis=1)
```
