### Let's observe which features have a strong relationship with the survival rate.
The first feature we will look at is 'Pclass'. <br>
We can expect that the high-class passengers will have a higher survival rate than lower-class passengers.<br>
Let's compare the proportions of survivals of each class.

```python
for i in range(1,4):
    print('Survival rate of class', i, 'is:')
    temp = train_data.loc[train_data['Pclass'] == i]['Survived'].value_counts()
    # Using loc fuction you can do 'slicing' with the specific column and 
    # the specific value.
    
    # when a column is sliced, type of the returned data is 'Series'.
    # 'Series' data has the function 'value_count' which will show the frequency
    # of each unique value in that series.
    
    print(temp[1]/temp.sum())
    print()
```

Our expectation that higher class passenger's survival rate is higher than others is correct.

The second feature we have to look at is 'Age'.<br>
We can expect that younger people's survival rate is higher than older people.
However, we have to categorize age data because, for example, it is hard to expect that age 28 and age 28.3 will have different survival rates. Therefore, I will make age groups (10 years interval) to categorize the age data we have for now.


+ 궁금한 점: 왜 나이를 그룹화 시켜야 하는가??

```python
# 0~10 group1, 10~20 group2, 20~30 group3, 30~40 group4,
# 40~50  group5, 50~60 group6, 60~70 group7, 70~ group8
age_group_list=[]
for i in range(len(train_data['Age'])):
    if 0<=train_data['Age'][i] and train_data['Age'][i]<=10:
        age_group_list.append(1)
    elif 10<train_data['Age'][i] and train_data['Age'][i]<=20:
        age_group_list.append(2)
    elif 20<train_data['Age'][i] and train_data['Age'][i]<=30:
        age_group_list.append(3)
    elif 30<train_data['Age'][i] and train_data['Age'][i]<=40:
        age_group_list.append(4)
    elif 40<train_data['Age'][i] and train_data['Age'][i]<=50:
        age_group_list.append(5)
    elif 50<train_data['Age'][i] and train_data['Age'][i]<=60:
        age_group_list.append(6)
    elif 60<train_data['Age'][i] and train_data['Age'][i]<=70:
        age_group_list.append(7)
    elif 60<train_data['Age'][i]:
        age_group_list.append(8)
```

```python
age_group = pd.DataFrame(age_group_list, columns=['age_group'])
```

```python
age_group
```

```python
train_data = pd.concat([train_data, age_group], axis=1)
# 'concat' function is for combining two different dataframes in terms of the index
# when we combine those horizontally. (axis=1)
```

```python
train_data
```

Let's see how survival rate is related with ages.

```python
for i in range(1,9):
    print('Survival rate of age group', i, 'is:')
    temp = train_data.loc[train_data['age_group'] == i]['Survived'].value_counts()
    print(temp[1]/temp.sum())
    print()
```

We can observe that age between 0 ~ 10 has the highest survival rate and passengers which have the age more than 60 have the lowest survival rate.
